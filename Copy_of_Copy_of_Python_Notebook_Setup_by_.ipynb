Research Question # 2 : Is there an association between importance in religious faith and total amount of unprescribed medication intake. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
!pip install ResearchPy
import researchpy as rp
```

```python
colors = sns.color_palette("Pastel2")
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```



Codebook 1: Screen Shot 2023-04-28 at 12.11.01 PM.png

```python
#Manage data for variable 1
df['H4RE9'].value_counts()
df['Religious_Importance'] = df['H4RE9']
df['Religious_Importance'].value_counts()
df['Religious_Importance'] = df['Religious_Importance'].loc[df['Religious_Importance'] !=' ']
df['Religious_Importance'] = df['Religious_Importance'].loc[df['Religious_Importance'] !='6']
df['Religious_Importance'] = df['Religious_Importance'].loc[df['Religious_Importance'] !='8']
df['Religious_Importance'].value_counts()
df['Religious_Importance'].replace(['1','2','3','4'], ['not important','somewhat important','very important ','more important than anything else'], inplace=True)
df['Religious_Importance'].value_counts()
```

```python
rp.summary_cat(df['Religious_Importance'])
```

```python
#Graphs and tables or statistics
plt.title('Average Religious Importancee')
df['Religious_Importance'].value_counts(normalize=True).plot(kind='barh', color=colors , width= 0.5, alpha = 0.5, grid=True)
```

```python
plt.title('Average Religious Importance')
df['Religious_Importance'].value_counts(normalize=True).plot(kind='pie', colors=colors)
plt.legend( loc='upper right', bbox_to_anchor=(1.85, 1))
```

```python
#Manage data for variable 1
df['H4TO63'].value_counts()
df['Unprescribed_Drugs'] = df['H4TO63']
df['Unprescribed_Drugs'].value_counts()
df['Unprescribed_Drugs'] = df['Unprescribed_Drugs'].loc[df['Unprescribed_Drugs'] !=' ']
df['Unprescribed_Drugs'] = df['Unprescribed_Drugs'].loc[df['Unprescribed_Drugs'] !='6']
df['Unprescribed_Drugs'] = df['Unprescribed_Drugs'].loc[df['Unprescribed_Drugs'] !='8']
df['Unprescribed_Drugs'].value_counts()
df['Unprescribed_Drugs'].replace(['0','1'], ['no','yes'], inplace=True)
df['Unprescribed_Drugs'].value_counts()
```

```python
rp.summary_cat(df['Unprescribed_Drugs'])
```

```python
#Graphs and tables or statistics
plt.title('Average Unprescribed Drug Intake')
df['Unprescribed_Drugs'].value_counts(normalize=True).plot(kind='barh', color=colors, width= 0.5, alpha = 0.5, grid=True)
```

```python
plt.title('Average Unprescribed Drug Intake')
df['Unprescribed_Drugs'].value_counts(normalize=True).plot(kind='pie', colors=colors)
plt.legend( loc='upper right', bbox_to_anchor=(1, 1))
```
